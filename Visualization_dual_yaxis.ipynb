```
import random

def generate_marketing_roi_data(num_campaigns):
    """
    Generates a random dataset of marketing campaign costs and resulting revenue.
    
    Args:
    - num_campaigns (int): The number of marketing campaigns to generate data for.
    
    Returns:
    - A tuple of two NumPy arrays, where the first array contains the cost of each campaign and the second
      array contains the resulting revenue.
    """
    # Generate random campaign costs between $1000 and $5000
    campaign_costs = np.array([random.randint(1000, 5000) for i in range(num_campaigns)])
    
    # Generate random revenue between $0 and 2 times the campaign cost
    campaign_revenue = np.array([random.uniform(0, 2) * cost for cost in campaign_costs])
    
    return campaign_costs, campaign_revenue
```

```
import numpy as np

num_campaigns = 10
campaign_costs, campaign_revenue = generate_marketing_roi_data(num_campaigns)

print(campaign_costs)
print(campaign_revenue)
```

```
import matplotlib.pyplot as plt
import numpy as np

def plot_roi_analysis(campaign_costs, campaign_revenue):
    """
    Plots a Dual Y-Axis Chart of marketing campaign costs and resulting revenue.
    
    Args:
    - campaign_costs (array): A NumPy array containing the cost of each marketing campaign.
    - campaign_revenue (array): A NumPy array containing the resulting revenue of each marketing campaign.
    
    Returns:
    - None
    """
    fig, ax1 = plt.subplots(figsize=(12, 8))

    # Plot campaign costs on the left y-axis
    ax1.plot(campaign_costs, color='tab:blue')
    ax1.set_xlabel('Campaign Number')
    ax1.set_ylabel('Campaign Cost ($)', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    # Create a second y-axis on the right for campaign revenue
    ax2 = ax1.twinx()

    # Plot campaign revenue on the right y-axis
    ax2.plot(campaign_revenue, color='tab:orange')
    ax2.set_ylabel('Campaign Revenue ($)', color='tab:orange')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    # Add a title and display the chart
    plt.title('Marketing ROI Analysis')
    plt.show()
```

```
plot_roi_analysis(campaign_costs, campaign_revenue)
```

```
import matplotlib.pyplot as plt
import numpy as np

def plot_roi_analysis(campaign_costs, campaign_revenue):
    """
    Plots a Dual Y-Axis Chart of marketing campaign costs and resulting revenue.
    
    Args:
    - campaign_costs (array): A NumPy array containing the cost of each marketing campaign.
    - campaign_revenue (array): A NumPy array containing the resulting revenue of each marketing campaign.
    
    Returns:
    - None
    """
    fig, ax1 = plt.subplots(figsize=(12, 8))

    # Plot campaign costs on the left y-axis
    ax1.plot(campaign_costs, color='tab:blue')
    ax1.set_xlabel('Campaign Number')
    ax1.set_ylabel('Campaign Cost ($)', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    # Create a second y-axis on the right for campaign revenue
    ax2 = ax1.twinx()

    # Plot campaign revenue on the right y-axis as a bar chart with alpha=0.6
    ax2.bar(np.arange(len(campaign_revenue)), campaign_revenue, color='tab:orange', alpha=0.6)
    ax2.set_ylabel('Campaign Revenue ($)', color='tab:orange')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    # Add a title and display the chart
    plt.title('Marketing ROI Analysis')
    plt.show()
```

```
plot_roi_analysis(campaign_costs, campaign_revenue)
```
